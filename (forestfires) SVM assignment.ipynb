#preprocessing:

```python
from sklearn import preprocessing
```

```python
labelencoder=preprocessing.LabelEncoder()
dataset["month"]=labelencoder.fit_transform(dataset["month"])
dataset["day"]=labelencoder.fit_transform(dataset["day"])
dataset["size_category"]=labelencoder.fit_transform(dataset["size_category"])
```

```python
dataset
```

```python
X=dataset.iloc[:,:-1]
```

```python
X
```

```python
Y=dataset.iloc[:,-1]
```

```python
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn import svm
from sklearn.svm import SVC
model=SVC()
param_grid=[{"kernel":["rbf"],"gamma":[50,25,20,15,5,9,2],"C":[15,10,5,2,1,0.01,0.001]}]
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(model,param_grid,cv=30)
grid.fit(x_train,y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
model1=SVC(gamma=50,C=15)
model1.fit(x_train,y_train)
```

```python
ptredict=model1.predict(x_test)
```

```python
print(classification_report(ptredict,y_test))
```

```python
print(confusion_matrix(ptredict,y_test))
```
