```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import seaborn as sns
import numpy as np
```

```python
wine=pd.read_csv('wine.csv')
wine.head()
```

```python
wine.shape
```

```python
data=wine.iloc[:,1:]
data
```

```python

array=data.values
array
```

```python
#Standarize:
standard=scale(array)
standard

```

```python
#PCA
pca=PCA(n_components=13)
pca_values=pca.fit_transform(standard)
pca_values
```

```python
#the amount of variance that each pca explains:
variance=pca.explained_variance_ratio_
```

```python
variance
```

```python
#comulative variance:
variance1=np.cumsum(np.round(variance,decimals=4)*100)
variance1
```

```python
pca.components_
```

```python
#PLOt
plt.plot(variance1,color="red")
```

```python
pcavalues=pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3'])

pcavalues
```

#kmeans:


```python
from sklearn.cluster import KMeans
```

```python
#Data Visualization:
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(pcavalues)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.title('Elbow Method')
plt.show()
```

```python
#Kmeans clustering model
new_clusters=KMeans(n_clusters=3,random_state=42)
new_clusters.fit(pcavalues)
```

```python
new_clusters.labels_
```

```python
#Conversion into dataframe
c1=pd.DataFrame(new_clusters.labels_,columns=['km_clust_id'])
```

```python
c1
```

```python
tm=data.copy(deep = True)
abc=pd.concat([c1,tm],axis=1)
abc
```

```python
abc.groupby(abc['km_clust_id']).mean()
```

```python
#Hierarachichal Cluatering:

from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
#Creating dendrogram using different method:
dendrogram=sch.dendrogram(sch.linkage(pcavalues,method='single'))
```

```python
dedrogram=sch.dendrogram(sch.linkage(pcavalues,method='complete'))
```

```python
h_clusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
ky=h_clusters.fit_predict(pcavalues)
ky
```

```python
c2=pd.DataFrame(ky,columns=['hir_clust_id'])
c2
```

```python
#join
po=pd.concat([c2,abc],axis=1)
po
```
